## Question 3: 
Create a bar chart that shows the frequency of diamonds grouped by cut. Explain the chart.


```python
import matplotlib.pyplot as plt

# Create the bar chart
plt.figure(figsize=(8, 6))
diamonds_df['cut'].value_counts().plot(kind='bar', color='skyblue')

# Customize the chart
plt.title('Frequency of Diamonds Grouped by Cut')
plt.xlabel('Cut')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
```

## Question 4: 
Create a scatterplot that shows the relationship between carat and price. Explain the chart and comment on the relationship between the variables.


```python
# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(diamonds_df['carat'], diamonds_df['price'], color='skyblue', alpha=0.5)

# Customize the chart
plt.title('Relationship between Carat and Price')
plt.xlabel('Carat')
plt.ylabel('Price')
```

## Question 5: 
Calculate the Pearson Correlation coefficient of carat and price. Explain the results and discuss the strength of the correlation.


```python
from scipy.stats import pearsonr

# Calculate the Pearson correlation coefficient and its associated p-value
corr_coef, p_value = pearsonr(diamonds_df['carat'], diamonds_df['price'])

# Print the results
print("Pearson Correlation Coefficient between carat and price:", corr_coef)
print("p-value:", p_value)
```

## Question 6: 
Create a histogram (or boxplot) that shows the distribution of prices based on the quality of the cut. Explain the charts and skew.
Note: this question is asking you to show the distribution of diamond prices for each cut. There should be a separate histogram/boxplot for each cut.


```python
# Get unique cut categories
cut_categories = diamonds_df['cut'].unique()

# Create a separate histogram for each cut quality
plt.figure(figsize=(12, 8))
for cut_category in cut_categories:
    plt.hist(diamonds_df[diamonds_df['cut'] == cut_category]['price'],
             bins=30, alpha=0.5, label=cut_category)

# Customize the chart
plt.title('Distribution of Prices Based on Cut Quality')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend(title='Cut Quality')
```

## Question 7: 
Using a 2-sample t-test**, determine if there is a statistical difference between the price of diamonds with a cut that is considered:
'Good' vs 'Very Good'
'Premium' vs 'Ideal'


```python
from scipy.stats import ttest_ind

# Extract prices for diamonds with 'Good' and 'Very Good' cut qualities
good_prices = diamonds_df[diamonds_df['cut'] == 'Good']['price']
very_good_prices = diamonds_df[diamonds_df['cut'] == 'Very Good']['price']

# Perform a two-sample t-test for 'Good' vs 'Very Good' cut quality
t_statistic, p_value = ttest_ind(good_prices, very_good_prices)

# Print the results
print("Results 'Good' vs 'Very Good'")
print("t-statistic:", t_statistic)
print("p-value:", p_value)
if p_value < 0.05:
    print("Statistically significant difference in prices between 'Good' and 'Very Good'.")
else:
    print("No statistically significant difference in prices between 'Good' and 'Very Good'.")
```

```python
# Extract prices for diamonds with 'Premium' and 'Ideal' cut qualities
premium_prices = diamonds_df[diamonds_df['cut'] == 'Premium']['price']
ideal_prices = diamonds_df[diamonds_df['cut'] == 'Ideal']['price']

# Perform a two-sample t-test for 'Premium' vs 'Ideal' cut quality
t_statistic, p_value = ttest_ind(premium_prices, ideal_prices)

# Print the results
print("Results of 'Premium' vs 'Ideal'")
print("t-statistic:", t_statistic)
print("p-value:", p_value)
if p_value < 0.05:
    print("Statistically significant difference in prices between 'Premium' and 'Ideal'.")
else:
    print("No statistically significant difference in prices between 'Premium' and 'Ideal'.")
```
